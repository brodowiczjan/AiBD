***Zadanie 9:***

```python
staff_pd = pd.read_sql("select distinct address_id from staff",con=connection_sqlalchemy) # niezduplikowane ID adresów pracowników
country_pd = pd.read_sql("select country_id from country where country in ('Argentina', 'Spain')",con=connection_sqlalchemy) # ID Argentyny i Hiszpani
city_pd = pd.read_sql("select city_id from city where country_id in " + str(tuple(country_pd.country_id)),con=connection_sqlalchemy) # ID miast z tych dwóch krajów
address_pd = pd.read_sql("select address_id from address where city_id in " + str(tuple(city_pd.city_id)),con=connection_sqlalchemy) # ID adresów z miast z dwóch kajów
add = [i for i in list(address_pd.address_id) if i in list(staff_pd.address_id)] # porównanie dwóch list
print("W Argentynie lub Hiszpanii mieszka", len(add), "pracowników.")
```

***Zadanie 10:***
