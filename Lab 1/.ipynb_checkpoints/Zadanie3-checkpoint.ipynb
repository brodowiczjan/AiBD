```python
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = lambda a: a**2 + 5  # tworzę funkcje 

l1 = np.linspace(-1, 1) #
l2 = np.linspace(-6, 6) # tworzę trzy przedziały dla funkcji
l3 = np.linspace(0, 5)  #
```

***Pierwszy przedział:***

```python
y1 = x(l1) # wyliczam wartości dla 1 przedziału

plt.plot(l1, y1)                          # tworzę wykres
plt.title("Wykres funkcji f(x)=x^2+5")    # tytuł
plt.xlabel("OX")                          # opis osi X
plt.ylabel("OY")                          # opis osi Y
plt.legend(["f(x)"])                      # legenda
plt.grid()                                # siatka
plt.show()                                # wyświetlam wykres
```

***Drugi przedział:***

```python
y2 = x(l2) # wyliczam wartości dla 2 przedziału

plt.plot(l2, y2)                          # tworzę wykres
plt.title("Wykres funkcji f(x)=x^2+5")    # tytuł
plt.xlabel("OX")                          # opis osi X
plt.ylabel("OY")                          # opis osi Y
plt.legend(["f(x)"])                      # legenda
plt.grid()                                # siatka
plt.show()                                # wyświetlam wykres
```

***Trzeci przedział:***

```python
y3 = x(l3) # wyliczam wartości dla 3 przedziału

plt.plot(l3, y3)                          # tworzę wykres
plt.title("Wykres funkcji f(x)=x^2+5")    # tytuł
plt.xlabel("OX")                          # opis osi X
plt.ylabel("OY")                          # opis osi Y
plt.legend(["f(x)"])                      # legenda
plt.grid()                                # siatka
plt.show()                                # wyświetlam wykres
```
