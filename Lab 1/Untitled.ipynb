```python
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = lambda a: a**2 + 5

l1 = np.linspace(-1, 1)
l2 = np.linspace(-6, 6)
l3 = np.linspace(0, 5)
```

***Pierwszy przedzia≈Ç:***

```python
y1 = x(l1)

plt.plot(l1, y1)
plt.title("Wykres funkcji f(x)=x^2+5")
plt.xlabel("OX")
plt.ylabel("OY")
plt.legend(["f(x)"])
plt.grid()
plt.show()
```

***Drugi przedzia≈Ç:***

```python
y2 = x(l2)

plt.plot(l2, y2)
plt.title("Wykres funkcji f(x)=x^2+5")
plt.xlabel("OX")
plt.ylabel("OY")
plt.legend(["f(x)"])
plt.grid()
plt.show()
```

***Trzeci przedzia≈Ç:***

```python
y3 = x(l3)

plt.plot(l3, y3)
plt.title("Wykres funkcji f(x)=x^2+5")
plt.xlabel("OX")
plt.ylabel("OY")
plt.legend(["f(x)"])
plt.grid()
plt.show()
```
